### Extra File to just store some simple plots for the exam.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create dataframe with columns "accuracy", "seconds"
df = pd.DataFrame(columns=["accuracy", "seconds", "vocab_size"])
```

```python
df.loc[0] = [0.853, 983, 1000]
df.loc[1] = [0.8871, 1635, 2000]
df.loc[2] = [0.9013, 3282, 5000]
```

```python
df
```

```python
# plot seconds vs vocab_size
plt.plot(df["vocab_size"], df["seconds"], "o-")
plt.ylabel("Elapsed time (seconds)")
plt.xlabel("Vocabulary Size")
plt.title("Elapsed Time vs Vocabulary Size")
plt.show()
```

```python
# plot accuracy vs vocab_size
plt.plot(df["vocab_size"], df["accuracy"], "o-")
plt.xlabel("Vocabulary Size")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Vocabulary Size")
plt.yticks(np.arange(0.85, 0.9, 0.005))
plt.show()
```
